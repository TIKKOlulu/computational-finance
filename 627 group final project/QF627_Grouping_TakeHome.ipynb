```python
import random
import pandas as pd
```

```python
roster = pd.read_csv("/Users/lu/Desktop/qf627_g1_takehome.csv")
roster
```

```python
students = roster["name"].tolist()
```

```python
random.seed(241029)
```

```python
members_per_group = 4
```

```python
group = 1

for s in students[:]:
    if members_per_group == 4:
        print("***** Group {} *****".format(group) + "\n"
             ) 
        members_per_group = 0
        group += 1
    
    student = random.choice(students)
    
    print(student + "\n")
    
    members_per_group += 1
    
    students.remove(str(student)
                   )
```
